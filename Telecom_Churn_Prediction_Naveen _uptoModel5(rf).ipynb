## Tag Churner

Now tag the churned customers (churn=1, else 0) based on the fourth month as follows: Those who have not made any calls (either incoming or outgoing) AND have not used mobile internet even once in the churn phase. The attributes you need to use to tag churners are:
 - total_ic_mou_9
 - total_og_mou_9
 - vol_2g_mb_9
 - vol_3g_mb_9
 
 
After tagging churners, remove all the attributes corresponding to the churn phase (all attributes having  _9, etc. in their names).

```python
# creating separate dataframe churn and nonchurn customers

churned_customers = high_value_cust[ high_value_cust['churn'] == '1' ]
non_churned_customers = high_value_cust[ high_value_cust['churn'] == '0' ]
```

```python
# Creating the function for univarient analysis for numerical feature

def uni_analysis_numerical(features,  scale='linear'):    
    
    data = high_value_cust
    
    fig = plt.figure(figsize=(16,8))
    ax1 = fig.add_subplot(1,3,1)
    sns.boxplot(x='churn', y = features[0], data = data, ax=ax1)
    title = ''.join(features[0]) +' vs Churn'
    ax1.set(title=title)
    if scale == 'log' :
        plt.yscale('log')
        ax1.set(ylabel= features[0] + '(Log Scale)')
        
    ax2 = fig.add_subplot(1,3,2)
    sns.boxplot(x='churn', y = features[1], data = data, ax=ax2)
    title = ''.join(features[1]) +' vs Churn'
    ax2.set(title=title)
    if scale == 'log' :
        plt.yscale('log')
        ax2.set(ylabel= features[1] + '(Log Scale)')
    
    ax3 = fig.add_subplot(1,3,3)
    sns.boxplot(x='churn', y = features[2], data = data, ax=ax3)
    title = ''.join(features[2]) +' vs Churn'
    ax3.set(title=title)
    if scale == 'log' :
        plt.yscale('log')
        ax3.set(ylabel= features[2] + '(Log Scale)')
    
    # summary statistic
    
    print('churned Customers(Churn : 1)')
    print(tabulate(churned_customers[features].describe(), tablefmt='grid'))
    
    print('\n Non-churn Customers(Churn : 0)')
    print(tabulate(non_churned_customers[features].describe(), tablefmt='grid'))
```

```python
# Creating the function for Univarent analysis for categorical features

def uni_analysis_categorical(features): 
    
    data = high_value_cust
    
    fig = plt.figure(figsize=[16,4])
    
    ax1 = fig.add_subplot(1,3,1)
    sns.countplot(x=features[0],hue='churn',data=data, ax=ax1)
    title = features[0] + ' vs No of Churned Customers'
    ax1.set(title= title)
    ax1.legend(loc='upper right')

    
    ax2 = fig.add_subplot(1,3,2)
    sns.countplot(x=features[1],hue='churn',data=data, ax=ax2)
    title = features[1] + ' vs No of Churned Customers'
    ax2.set(title= title)
    ax2.legend(loc='upper right')

        
    ax3 = fig.add_subplot(1,3,3)
    sns.countplot(x=features[2],hue='churn',data=data, ax=ax3)
    title = features[2] + ' vs No of Churned Customers'
    ax3.set(title= title)
    ax3.legend(loc='upper right')

        
    # Percentages 
    print('Customers who churned (Churn : 1)')
    print(tabulate(pd.DataFrame(churned_customers.stb.freq([features[0]], thresh=97)), headers='keys', tablefmt='grid'),'\n')
    print(tabulate(pd.DataFrame(churned_customers.stb.freq([features[1]], thresh=97)), headers='keys', tablefmt='grid'),'\n')
    print(tabulate(pd.DataFrame(churned_customers.stb.freq([features[2]], thresh=97)), headers='keys', tablefmt='grid'),'\n')

    print('\nCustomers who did not churn (Churn : 0)')
    print(tabulate(pd.DataFrame(non_churned_customers.stb.freq([features[0]], thresh=97)), headers='keys', tablefmt='grid'),'\n')
    print(tabulate(pd.DataFrame(non_churned_customers.stb.freq([features[1]], thresh=97)), headers='keys', tablefmt='grid'),'\n')
    print(tabulate(pd.DataFrame(non_churned_customers.stb.freq([features[2]], thresh=97)), headers='keys', tablefmt='grid'),'\n')
```

```python
# Creating the function to calculate VIF

from statsmodels.stats.outliers_influence import variance_inflation_factor

def get_vif(cols):
    df = X_train_resampled[cols]
    vif = pd.DataFrame()
    vif['Features'] = df.columns
    vif['VIF'] = [ variance_inflation_factor(df.values, i) for i in range (df.shape[1]) ]
    vif['VIF'] = round(vif['VIF'], 2)
    return vif.sort_values(by= 'VIF', ascending= False )
```

#### a.ROC Curve

```python
# Creating the function to plot ROC curve

from sklearn import metrics

def plot_roc( actual, probs ):
    fpr, tpr, thresholds = metrics.roc_curve( actual, probs, drop_intermediate = False )
    auc_score = metrics.roc_auc_score( actual, probs )
    plt.figure(figsize=(6, 4))
    plt.plot( fpr, tpr, label='ROC curve (area = %0.2f)' % auc_score )
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver operating characteristic example')
    plt.legend(loc="lower right")
    plt.show()
```

```python
# Plotting the ROC curve for the model.

plot_roc(y_train_pred_df['Churned'].astype('int'), y_train_pred_df['Predicted_prob'])
```

### b. Trade off between metrics and choosing optimal cut-off

```python
# Creating the features with different cut-off probability

cut_offs = [float(i)/10 for i in range(10)]

for i in cut_offs:
  y_train_pred_df[i] = y_train_pred_df['Predicted_prob'].map(lambda x : 1 if x>i else 0)

y_train_pred_df.head()
```

```python
# Creating the evaluation metrics dataframe

metrics_df = pd.DataFrame( columns=['Cut-offs', 'Accuracy', 'Sensitivity', 'Specificity'] )

from sklearn.metrics import confusion_matrix

for i in cut_offs:
    c_matrix = confusion_matrix((y_train_pred_df['Churned']).astype('int'), y_train_pred_df[i])

    Accuracy = (c_matrix[0,0] + c_matrix[1,1]) / sum(sum(c_matrix))
    Sensitivity = c_matrix[1,1] / (c_matrix[1,0] + c_matrix[1,1])
    Specificity = c_matrix[0,0] / (c_matrix[0,0] + c_matrix[0,1])

    metrics_df.loc[i] = [i, Accuracy, Sensitivity, Specificity]

metrics_df.head()
```

```python
# Plotting the metrics

metrics_df.plot.line( y=['Accuracy','Sensitivity','Specificity'],figsize=(24,8), grid=True, xticks=np.arange(0,1,0.02),
                title='Specificity-Sensitivity TradeOff' )
```

- From the curve above 0.55 could be the optimal cut-off point.

```python
# Creating the new feature with optimal_cutoff predicted value

optimal_cutoff = 0.55

y_train_pred_df['churn_predicted'] = y_train_pred_df['Predicted_prob'].map(lambda x : 1 if x > optimal_cutoff else 0)

y_train_pred_df.head()
```

### Model Evaluation (Logistic Regression)

```python
# Creating the function for model evaluation

import math
def model_metrics(matrix) :
    TN = matrix[0][0]
    TP = matrix[1][1]
    FP = matrix[0][1]
    FN = matrix[1][0]
    accuracy = round((TP + TN)/float(TP+TN+FP+FN),3)
    print('Accuracy :' ,round(accuracy*100, 2), '%' )
    sensitivity = round(TP/float(FN + TP),3)
    print('Sensitivity / True Positive Rate / Recall :', round(sensitivity*100,2), '%')
    specificity = round(TN/float(TN + FP),3)
    print('Specificity / True Negative Rate : ', round(specificity*100,2), '%')
    precision = round(TP/float(TP + FP),3)
    print('Precision / Positive Predictive Value :',  round(precision*100,2), '%')
    print('F1-score :', round( (2*precision*sensitivity/(precision + sensitivity))*100 , 2), '%'  )
    
```
